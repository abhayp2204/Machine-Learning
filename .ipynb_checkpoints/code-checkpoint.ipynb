```python
import pickle
import numpy as np
import pandas as pd
from sklearn.model_selection import ShuffleSplit
```

```python
# Depickle
pickle_in = open("data/train.pkl", "rb")
result = pickle.load(pickle_in)
```

```python
# Convert to numpy array and display the first 5 elements
train_set_full = np.array(result)
train_set_full[:5]
```

```python
# Shuffle the dataset
np.random.shuffle(train_set_full)
train_set_full
```

```python
# Separate the dataset
n = 16
train_set = np.split(train_set_full, n)
train_set[0][:5]
```
