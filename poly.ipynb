```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Depickle
pickle_in = open('data/train.pkl', 'rb')
data = pickle.load(pickle_in)
```

```python
# Shuffle the dataset
# np.random.shuffle(data)
```

```python
# Separate the dataset
num_datasets = 16
plt.scatter(data[:,0], data[:, 1], s=10)
train_sets = np.split(data, num_datasets)
```

```python
# Setup
polynom = PolynomialFeatures(degree=100)
```

```python
for train_set in train_sets:    
    # Separate the features and labels
    instances = 500
    x = train_set[:instances, 0]
#     x = np.sort(x)
    y = train_set[:instances, 1]

    # Convert X from a 1d array to a 2d array
    x = x[:, np.newaxis]
    y = y[:, np.newaxis]
    print(x.shape)
    print(y.shape)

    poly = PolynomialFeatures(degree=4)
    x_poly = poly.fit_transform(x)

    model = LinearRegression()
    a = model.fit(x_poly, y)
    print(a.coef_)
    y_poly_pred = model.predict(x_poly)

    # Polynomial regression results
    plt.scatter(x, y, s=10)
    plt.plot(x, y_poly_pred, color='m')
    plt.show()
```
