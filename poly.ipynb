```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Depickle
pickle_in = open('data/train.pkl', 'rb')
result = pickle.load(pickle_in)
result[:1]
```

```python
# Shuffle the dataset
np.random.shuffle(result)
```

```python
# Separate the dataset
num_datasets = 16
train_sets = np.split(result, num_datasets)
```

```python
# Setup
polynom = PolynomialFeatures(degree=2)
```

```python
# Loop
for train_set in train_sets:
    polynom = PolynomialFeatures(degree=3)
    instances = 5
    # Separate the features and labels
    X = train_set[:instances, 0]
    y = train_set[:instances, 1]
    
    # Convert X from a 1d array to a 2d array
    shape = (len(X), 1)
    X = np.reshape(X, shape)
    
    # Split the dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    # Coefficients: ax2 + bx + c
    X_polynom = polynom.fit_transform(X_train)
#     print(X_polynom)
    
    # Fit the polynomial regression model
    PolyReg = LinearRegression()
    PolyReg.fit(X_polynom, y_train)
    predictions = PolyReg.predict(polynom.fit_transform(X_train))
    
    # Polynomial regression results
    plt.scatter(X, y, color='red')
    plt.plot(X_train, predictions, color='blue')
    plt.title('Polynomial Regression')
    plt.xlabel('Area')
    plt.ylabel('Cost')
    plt.show()
```
